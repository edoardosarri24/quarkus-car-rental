```python
import numpy as np
```

```python
data = np.random.exponential(3, size=1490)
```

```python
data = np.random.exponential(3, size=1490)
print(f"data: {data}")
outliers = np.random.normal(loc=100, scale=50, size=5)
print(f"outliers: {outliers}")
```

```python
print("-Whitout outliers:")
print(f"Mean:{np.mean(data)}")
print(f"Std:{np.std(data)}")
print(f"CV:{np.std(data)/np.mean(data)}")

data_with_outliers = np.hstack((data, outliers))

print("-With outliers:")
print(f"Mean:{np.mean(data_with_outliers)}")
print(f"Std:{np.std(data_with_outliers)}")
print(f"CV:{np.std(data_with_outliers)/np.mean(data_with_outliers)}")
```
